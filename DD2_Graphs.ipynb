#Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

#The first graph ( HPWL vs Temperature)

##d0

```python
d0_df=pd.read_csv('annealing_datad0.csv')
```

```python
d0_df
```

```python
fig = px.line(d0_df
              , x='CurrentTemperature', y='TotalHPWL', title='Annealing Progress')
```

```python
fig = px.line(d0_df, x='CurrentTemperature', y='TotalHPWL', title='Annealing Progress',line_shape='linear')
interpolated_line = np.linspace( d0_df['TotalHPWL'].iloc[0],d0_df['TotalHPWL'].iloc[-1], len(d0_df))
fig.add_scatter(x=d0_df['CurrentTemperature'], y=interpolated_line, mode='lines', name='Interpolated Line')
fig.update_layout(
    xaxis_title='Current Temperature',
    yaxis_title='Total HPWL',
    showlegend=True
)
fig.update_xaxes(autorange="reversed")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


plt.plot(d0_df['CurrentTemperature'], d0_df['TotalHPWL'], label='Annealing Progress')

interpolated_line = np.linspace(d0_df['TotalHPWL'].iloc[0], d0_df['TotalHPWL'].iloc[-1], len(d0_df))
plt.plot(d0_df['CurrentTemperature'], interpolated_line, label='Interpolated Line', linestyle='--')

plt.title('Annealing Progress')
plt.xlabel('Current Temperature')
plt.ylabel('Total HPWL')
plt.legend()
plt.gca().invert_xaxis()
plt.xscale('log')
plt.show()
```

```python
fig.show()
```
